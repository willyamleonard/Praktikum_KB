### grafik batang

```python
dfjumlahType= df['Type'].value_counts()     

plt.figure(figsize=(6,3))

plt.bar(dfjumlahType.index, dfjumlahType,color = ['blue','orange'])
plt.xlabel('Type')
plt.ylabel('Type')
plt.title('perbandingan Type',fontsize=20,fontweight='bold')
```

##### range L dibawah 6000 dan range M dibawah 3000 dan raange H diantara 1000 yang mana range L lebih banyak 

### Histogram

```python
plt.figure(figsize=(6,3))
plt.hist(df['Type'],color='green')

plt.xlabel('jenis Type')
plt.ylabel('jumlah Type')
plt.title('perbandingan jenis Type',fontsize=10,fontweight='bold')
plt.show()
```

##### Range M sampai 3000 Range L di bawah 6000 dan H dibawah 2000 dan Dari range itu yang Paling banyak adalah L

#### Pie Charts

```python
dfjumlahType = df['Type'].value_counts()
dfjumlahType
```

```python
plt.pie(dfjumlahType, 
        labels = dfjumlahType.index, 
        autopct="%1.1f%%", 
        explode=[0,0,0.1])

plt.show()
```

##### Range M sampai 30% Range L di 60% dan H di 10% dan Dari range itu yang Paling banyak adalah L dengan menempati presentase tertinggi

### Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_iris.corr()
```

```python
plt.figure(figsize=(12,6))

sns.heatmap(
    df_iris.corr(),
    annot=True,
    cmap='flare'
)
```

##### kesimpulan dari heatmap yakni nilai tengah nya di mulai dari machine failure yang dimana semakin kebawah semakin besar dan semakin keatas semakin besar juga
