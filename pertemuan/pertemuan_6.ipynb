```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df =pd.read_csv('breast-cancer.csv')
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
x= df.drop('irradiat',axis=1).copy()
y= df['irradiat']
```

```python
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()

y = lbl.fit_transform(y)
```

```python
x = pd.get_dummies(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    x, y, test_size=0.2, random_state=30
)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler() 
```

```python
x_train
```

```python
x_scaled_train = ss.fit_transform(x_train)
x_scaled_test = ss.fit_transform(x_test)
```

```python
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

# input layer
model.add(Dense(16,activation = 'relu', input_dim=41))

# hidden
model.add(Dense(32, activation='relu'))
model.add(Dense(64, activation='relu'))

# output
model.add(Dense(1, activation='sigmoid'))

from keras.callbacks import EarlyStopping
```

```python
monitor_callbacks = EarlyStopping(monitor='accuracy', patience=5)
```

```python
from keras.optimizers import Adam
from keras.losses import binary_crossentropy

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_scaled_train, y_train,
          validation_data=(x_scaled_test,y_test),
          epochs=95,
          verbose=1,
          callbacks=[monitor_callbacks])
```

```python
model.evaluate(x_scaled_test, y_test)
```
