```python
import seaborn as sns

df = sns.load_dataset('titanic')
df.head()
```

```python
dfjumlahsex = df['sex'].value_counts()     

plt.figure(figsize=(6,3))

plt.bar(dfjumlahsex.index, dfjumlahsex,color = ['blue','orange'])
plt.xlabel('jenis kelamin')
plt.ylabel('jumlah kelamin')
plt.title('perbandingan jenis kelamin',fontsize=20,fontweight='bold')
```

```python
dfjumlahsex = df['sex'].value_counts()

plt.figure(figsize=(6,3))

plt.barh(dfjumlahsex.index, dfjumlahsex,color = ['blue','orange'])
plt.xlabel('jenis kelamin')
plt.ylabel('jumlah kelamin')
plt.title('perbandingan jenis kelamin',fontsize=20,fontweight='bold')
```

##### seaborn

```python
sns.set_style('whitegrid')
sns.countplot(data=df,x='class', hue='sex')
```

### histogram

```python
plt.figure(figsize=(6,3))
plt.hist(df['fare'],color='green')

plt.xlabel('jenis kelamin')
plt.ylabel('jumlah kelamin')
plt.title('perbandingan jenis kelamin',fontsize=20,fontweight='bold')
plt.show()
```

### pie chart

```python
dfwhocounts = df['who'].value_counts()
dfwhocounts
```

```python
plt.pie(dfwhocounts, 
        labels = dfwhocounts.index, 
        autopct="%1.1f%%", 
        explode=[0,0,0.1])

plt.show()
```

### scatter plot

```python
df_iris = sns.load_dataset('Iris')
df_iris.head()
```

```python
sns.scatterplot(data=df_iris,
             x = "sepal_length",
             y = "petal_length",
             hue = 'species'
             )

plt.show()
```

#### boxplot

```python
sns.boxplot(
    data = df_iris,
    x = "species",
    y = "sepal_length"
)

plt.show()
```

### heatmap

```python
df_iris.corr()
```

```python
plt.figure(figsize=(6,3))

sns.heatmap(
    df_iris.corr(),
    annot=True,
    cmap='flare'
)
```
